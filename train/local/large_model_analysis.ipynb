```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
f=open("large_model_result.txt",'r')
train_log=[]
test_log=[]
for line in f.readlines():
    mod, val = line.rstrip().split()
    if mod == "train:":
        train_log.append(float(val))
    else:
        test_log.append(float(val))
plt.figure(figsize=(10,4))
plt.plot(np.arange(len(train_log))*5, train_log)
plt.plot([0, len(train_log)*5], [0.585,0.585],'--')
plt.plot(np.arange(len(test_log))*5,test_log)
plt.plot([0, len(test_log)*5], [0.247,0.247],'--')
plt.legend(["train","train_baseline","legendary","test_baseline"], loc="best")
plt.ylabel("classification rate")
plt.xlabel("iteration")
plt.savefig("large_model_analysis.jpg")
```
